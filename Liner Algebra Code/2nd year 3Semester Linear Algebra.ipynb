three rules for multiplatin @ in a matrix 

```python
import numpy as np
import matplotlib.pyplot as plt
# if @>1
alpha=3
v=np.array([33,44,5,6])
result=alpha*v
print("first graph  @>1 \n")
plt.plot([0,v[0]],[0,v[1]])
plt.axis([-20,20,-30,30])

```

```python
import numpy as np
import matplotlib.pyplot as plt
# if @<0
alpha=-6
v=np.array([33,44,5,6])
result=alpha*v
print("Second graph  @<0 \n")
plt.plot([0,v[0]],[0,v[1]])
plt.axis([-20,20,-30,30])
```

if 0<@<1

```python
import numpy as np
import matplotlib.pyplot as plt
# if 0<@<1
alpha=0.5
v=np.array([33,44,5,6])
result=alpha*v
print("Third graph  0<@<1 \n")
plt.plot([0,v[0]],[0,v[1]])
plt.axis([-20,20,-30,30])
```

Code Challing    
Creat a 2D vector v, 10 Scalars that are drawn at random from a normal (Gaussiau ) distribution and plot all 10 scalar vector multiplication on top of each other.  
What do you notice ?

```python
import numpy as num
import  matplotlib.pyplot as mplt
v=num.array([1,2])
mplt.plot([0,v[0]],[0,v[1]])
for i in range (10):
  s=num.random.randn()
sv=s*v
mplt.plot([0,sv[0]],[0,sv[1]])
mplt.grid('on')
mplt.axis([-4,4,-4,4])
```
